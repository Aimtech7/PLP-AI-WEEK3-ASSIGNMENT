# spaCy NER + Rule-based Sentiment Starter

```python
import spacy
from textblob import TextBlob
nlp = spacy.load('en_core_web_sm')
text = 'I love the Samsung Galaxy S23 camera but the battery life is disappointing.'
doc = nlp(text)
print('Entities:')
for ent in doc.ents:
    print(ent.text, ent.label_)
pol = TextBlob(text).sentiment.polarity
print('Sentiment polarity:', pol)
```
